```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, [2]*len(x), linestyle='solid')
plt.show()
```

```python
plt.plot(x, 2 * x, linestyle='solid')
plt.show()
```

```python
plt.plot(x, 2 * x + 3, linestyle='solid', color='green')
plt.show()
```

```python
# Non-linear functions
plt.plot(x, np.sin(x), linestyle='solid', color='red')
plt.show()
```

```python
a = 1
b = 0
p = 3

# Creating the x and y vectors
x = np.linspace(-1, 1, 100)
y = a * x ** p + b

# Choosing the figure size
plt.figure(figsize=(6, 6))

ax = plt.gca()
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.5))

# Drawing axes
plt.axhline(0, c='#A9A9A9')
plt.axvline(0, c='#A9A9A9')

plt.axis('equal')

plt.plot(x, y)
plt.show()
```
